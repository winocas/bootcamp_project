NA_Sales 중 K와 M이 포함된 데이터들은  
K가 있을 경우 이를 원 데이터의 단위에 맞추기 위해 K만 떼고 0.001을 곱해줘야했으며  
M이 있을 경우는 단순히 M을 떼기만 하면 된다. 

```
df2[df2[sales_var[0]].str.contains('K|M')][sales_var[0]].replace({'K': '*1e-03', 'M': ''},regex=True).map(pd.eval).astype(float)
```

M이 있던 데이터들은 M만 사라지고  
K가 있던 데이터들은 K가 사라지고 0.001이 각각 곱해졌음을 알 수 있다. 

한 열에 대해서 처리를 해봤고 이를 반복문을 통해서 실제 데이터 변수들을 변환해주면

```
df3 = df2.copy() # 데이터 달라질테니 복사
```

```
sales_var
```

```
NA_index = df3.loc[df3.loc[:,'NA_Sales'].str.contains('K|M')]['NA_Sales'].index
```

반복문을 통해 한번에 수행해보면

```
for i,k in enumerate(sales_var):
    index= df3.loc[df3.loc[:,k].str.contains('K|M')][k].index
    df3.loc[index,k] = df3[df3[sales_var[i]].str.contains('K|M')][sales_var[i]].replace({'K': '*1e-03', 'M': ''},regex=True).map(pd.eval).astype(float)
```

```
df3[sales_var].apply(lambda x : x.str.contains('K|M')).sum() # K또는 M이 없다
```
